```python
import numpy as np
import math
from mealpy import FloatVar, BFO

global count
count = 0

def Rosenbrock(x):
    f = []
    global count
    count += 1
    for i in range(len(x) - 1):
        f.append(100 * (x[i] ** 2 - x[i + 1]) ** 2 + (x[i] - 1) ** 2)
    return np.sum(f)

def Rastrigin(x):
    f = []
    global count
    count += 1
    for i in range(len(x)):
        z_i = 1 + (x[i] + 1) / 4
        f.append(z_i ** 2 - 10 * math.cos(2 * math.pi * z_i) + 10)
    return np.sum(f)

def Weierstrass(x, a=0.5, b=3):
    global count
    k_max = 20
    f = []
    count += 1
    for k in range(len(x)):
        termo = 0
        for i in range(k_max):
            termo += a ** k * np.cos(2 * np.pi * b ** k * (x[i] + 0.5))
        f.append(termo)
    return np.sum(f)


D = 10

problem_dict = {
    "bounds": FloatVar(lb=[-100, ] * 10, ub=[100, ] * 10,),
    "obj_func": Rastrigin,
    "minmax": "min",
    "log_to": None
}

term_dict = {
    "max_fe": 10000 * D
}

listaResult = []
evaluations = []

for i in range (51):
    model = BFO.ABFO(epoch=500, pop_size=200, n_limits = 50)
    g_best = model.solve(problem_dict, termination=term_dict)
    listaResult.append(g_best.target.fitness)
    evaluations.append(count)  # Store the count for this run
    print(f"Execução {i+1}: Melhor fitness = {g_best.target.fitness}, Avaliações = {count}")

# Calcular estatísticas
mean_evaluations = np.mean(evaluations)
std_evaluations = np.std(evaluations)

# Imprimir resultados formatados
print(f"Lista de Resultados: {listaResult}")
print(f"Lista de Avaliações: {evaluations}")
print(f"Avaliações Totais: {sum(evaluations)}")
print(f"Precisão: {mean_evaluations:.2f} ± {std_evaluations:.2f}")
print(f"Melhor Fitness: {min(listaResult)}")
print(f"Fitness Médio: {np.mean(listaResult)}")
print(f"Mediana do Fitness: {np.median(listaResult)}")
print(f"Pior Fitness: {max(listaResult)}")
print(f"Desvio Padrão do Fitness: {np.std(listaResult)}")

# Algo Começou a dar errado, e não sei exatamento oq é, preciso corrigir;
```
